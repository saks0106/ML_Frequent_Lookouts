**Feature Construction -** Creating Features by Merging 2 or more columns. Here Domain and Field Knowledge plays an important role

**Feature Splitting -** Numerical or/and Categorical are merged together, so we do feature splitting. Eg - Mr ABC should be split into Mr. and ABC

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

## Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1 # Sibling/Spouse + Parent/Child + Passenger
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

## Feature Splitting

```python
df = pd.read_csv('train.csv')
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
(df.groupby('Title').mean()['Survived']).sort_values(ascending=False)
```
