```python
#using matplotlib method
np.random.seed(42)
plt.figure(figsize=(10,6))
months = np.arange(1,11,1)
adcost = np.array([1,56,3,78,56,34,23,98,2,51])
sales = np.random.randint(1,200,size=10)
revenue = np.random.randint(1,50,size=10)
plt.xlabel("Months")
plt.ylabel('Sales and AdCost')
plt.grid()
plt.plot(months,adcost,marker='o',c='cyan',linestyle='dashed');
plt.plot(months,sales,marker='^',c='orange',linewidth=5,alpha=0.5);
plt.bar(months,revenue,color='lightgreen')
plt.legend(['Adcost','Sales','Revenue']);
for i in range(len(months)):
    plt.text(x=months[i],y=adcost[i],s=adcost[i],fontsize=8,bbox=dict(facecolor='red', alpha=0.8))
    plt.text(x=months[i],y=sales[i],s=sales[i],fontsize=8,bbox=dict(facecolor='blue', alpha=0.2))
    plt.text(x=months[i],y=revenue[i],s= revenue[i],fontsize=8,bbox=dict(facecolor='lightblue', alpha=0.2))
plt.xticks(ticks=months,labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct']);
```

```python
fig,ax = plt.subplots(2, 2, sharex='all', sharey='col')
```

```python
#OO method
fig,ax = plt.subplots(num=10,figsize=(10,6))
ax.set(title='SaGo',xlabel='Months',ylabel='Revenue')
ax.grid()
ax.plot(months,sales,color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
ax.plot(months,adcost,color='salmon', marker='*', linestyle='-',linewidth=2, markersize=12)
ax.bar(months,revenue,color='skyblue')
ax.legend(['sales','adcost','revenue'])
#for plotting values on the graph
for i in range(len(months)):
    ax.text(months[i],sales[i],s=sales[i])
    ax.text(months[i],adcost[i],s=adcost[i])
    ax.text(months[i],revenue[i],s=revenue[i])
ax.set_xticks(ticks=months
              ,labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct']);
```

```python
df = pd.read_csv('/content/heart-disease.csv')
over50 = df[df['age']>50]
over50
```

```python
#heart disease with details
#sumplot of chol, age, thalach(max heart rate)
fig, (ax0,ax1) = plt.subplots(nrows= 2,ncols=1, figsize=(10,10), sharex= True)# sharex is common x axis, we can remove it in ax0
#add ax0
scatter = ax0.scatter(x= over50['age'], y= over50['chol'],c= over50['target'], cmap='summer')
#customize ax0
ax0.set(title= 'Heart Disease and Cholestrol Levels',ylabel= 'cholestrol')
#change x axis limit, the extra space on left side, will work for both subplots as they share same axis
ax0.set_xlim([50,80])
#add a legend to ax0
ax0.legend(*scatter.legend_elements(),title= "Target")
#add a mean line
ax0.axhline(y=over50['chol'].mean(),linestyle='--');
#add data to ax1
scatter = ax1.scatter(x= over50['age'], y= over50['thalach'], c=over50['target'],cmap='winter');
#customize ax1
ax1.set(title='Heart Disease and Max heart rate', xlabel='age',ylabel='max heart rate');
#change x axis limit. AGAIN THIS IS OPTIONAL
ax1.set_xlim([50,80])
ax1.set_ylim([60,200]) #this is not optional for chances are needed for y axis
#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");
#add  a mean line
ax1.axhline(y=over50['thalach'].mean(),linestyle='--');
#add a title to entire figure
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight='bold');
```
