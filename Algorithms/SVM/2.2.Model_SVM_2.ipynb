SVM Kernels will do transformation from low dims to high dims. 

Linear kernel - will draw a straight line across data points

```python
#Import base libraies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.svm import SVC
```

```python
df = pd.read_csv('/content/pima-indians-diabetes-2.csv')
df.head()
```

```python
X = df.drop('class',axis=1)
y = df['class']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x='pedi',y='skin',z='mass',color='class')
fig.show()
```

```python
#Linear Kernel
model = SVC(kernel='linear')
model.fit(X_train,y_train)

from sklearn.metrics import accuracy_score
y_preds1 = model.predict(X_test)
accuracy_score(y_test,y_preds1)
```

```python
#Polynomial kernal
from sklearn.svm import SVC
poly = SVC(kernel='poly')
poly.fit(X_train,y_train)

y_preds2 = poly.predict(X_test)
accuracy_score(y_test,y_preds2)
```

```python
#rbf kernal
from sklearn.svm import SVC
rbf = SVC(kernel='rbf') # radial basis function
rbf.fit(X_train,y_train)

y_preds3 = rbf.predict(X_test)
accuracy_score(y_test,y_preds3)
```

```python
#sigmoid kernal
from sklearn.svm import SVC
sig = SVC(kernel='sigmoid')
sig.fit(X_train,y_train)

y_preds = sig.predict(X_test)
accuracy_score(y_test,y_preds)
```

```python
#HyperTuning using GridSearchCV
from sklearn.model_selection import GridSearchCV

params = [{'C':[0.1,0.2,0.4,0.6,0.8,1],
          'kernel':['linear','poly','rbf','sigmoid'],
           'gamma':['scale','auto']}]

model = SVC()
gsmodel = GridSearchCV(model,param_grid=params,scoring='accuracy',cv=10)
gsmodel.fit(X_train,y_train)
```
